```python
from third_party.LDL.code import train
```

```python
cnn = train.trainval_test(cross_val_index='0', sigma=3.0, lam=0.6)
```

```python
from torchvision import transforms
from third_party.LDL.code.dataset import dataset_processing
from torch.utils.data import DataLoader
import torch
```

```python
TEST_FILE = '../Classification/NNEW_test_' + str(0) + '.txt'
DATA_PATH = '../Classification/JPEGImages'

normalize = transforms.Normalize(mean=[0.45815152, 0.361242, 0.29348266],
                                     std=[0.2814769, 0.226306, 0.20132513])
```

```python
dset_test = dataset_processing.DatasetProcessing(
        DATA_PATH, TEST_FILE, transform=transforms.Compose([
                transforms.Scale((224, 224)),
                transforms.ToTensor(),
                normalize,
            ]))

test_loader = DataLoader(dset_test,
                        batch_size=10,
                        shuffle=False,
                        num_workers=1,
                        pin_memory=True)
```

```python
test_img, y, l = next(iter(test_loader))
```

```python
preds = cnn(test_img.cuda(), None)
```

```python
_, preds_cls = torch.max(preds[0] + preds[2], 1)
_, preds_cnt = torch.max(preds[1], 1)
preds_cnt = (preds_cnt + 1)
```

```python
print('true cls: ', y)
print('pred cls: ', preds_cls)
```

```python
print('true cnt: ', l)
print('pred cls: ', preds_cnt)
```
